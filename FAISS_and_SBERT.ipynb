```python
from sentence_transformers import SentenceTransformer

# Load the pre-trained model
# model = SentenceTransformer('all-MiniLM-L6-v2')
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Example sentences
sentences = ["Hello This is Akash", "Together we'll shine", "Mahir Tajuar Akash", " Hello everyone This is Akash"]

# Generate embeddings for the sentences
embeddings = model.encode(sentences)

# Print embeddings
print(embeddings)
```

```python
#Shape of Sentence Embedding

embeddings.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

#loop through to get the similary indexes

score = np.zeros( (embeddings.shape[0] , embeddings.shape[0]) )  # An array of 3*3 dimension  , np.zeros means all values zero

for i in range(embeddings.shape[0]):
  score[i,:] = cosine_similarity(
      [embeddings[i]],
      embeddings
  )
```

An index in FAISS is like a big **book of labels** that helps you **find things faster**. Imagine you have a **big toy box** full of toys. Instead of looking at every toy one by one, you have a list (like a label) that tells you where each toy is, so you can find them much faster!

In details

In FAISS, an index is a **data structure** designed to store vectors (or embeddings) and allows for **efficient similarity search**. When working with high-dimensional data, an index organizes the vectors in a way that minimizes the number of comparisons needed during a query. By partitioning the data or using methods like **Product Quantization** or **IVF** (Inverted File Index), FAISS can speed up the search process by quickly narrowing down the potential matches, making it suitable for handling large-scale datasets.

"Quantization" refers to the process of mapping a large range of values (such as continuous data) into a smaller set of discrete values.

In the context of **Product Quantization (PQ)** in FAISS:

- **Quantization** means converting continuous, high-dimensional vectors into a set of smaller, discrete values that represent the most important features of the original data.
- This is achieved by dividing the original vector into smaller parts (subvectors) and then finding the closest centroid (a representative value) from a predefined set of centroids (codebook) for each subvector.

The result is a more compact representation of the original vector, which saves memory and speeds up similarity searches, though it comes at the cost of some approximation.

In FAISS, an index is a data structure used to store and organize vectors (such as sentence embeddings) in a way that allows for fast and efficient similarity searches. Without an index, you would have to compare every single vector with the query vector, which would be very slow, especially as the dataset grows.

<br>
 Compatibility:
If you try to add vectors with different dimensions to the same FAISS index, it would cause an error because FAISS expects all vectors in the same index to have the same number of features (dimension).
By specifying the dimension (dim) when creating the index, you ensure that FAISS will correctly handle and organize the vectors in the index.

### 2. `D, I`
- **`D`** is a matrix (or array) of **distances**. It contains the distances between the query vector and the retrieved vectors (nearest neighbors). In most cases, it will be the squared Euclidean distance or cosine distance, depending on the type of index you are using.
  - The closer the distance, the more similar the vectors are.
  - `D` will have shape `(1, k)`, where `k` is the number of nearest neighbors you requested, and it contains the actual distance values.

- **`I`** is a matrix (or array) of **indices**. It contains the **indices of the nearest vectors** in the FAISS index. These indices point to the positions of the closest vectors in the original index, which can be used to look up the actual data associated with those vectors.
  - `I` will have shape `(1, k)` where `k` is the number of nearest neighbors.
  - The values in `I` correspond to the indices of the nearest neighbors in the FAISS index.

### Example:

Let's say we have a set of vectors in the FAISS index (representing sentences, for example), and we want to find the 3 closest sentences to a new query sentence.

- **`query`** is the query vector, the embedding of the new sentence.
- **`k = 3`** is the number of nearest neighbors to search for.

After executing `D, I = index.search(query, 3)`, we will get:

- **`D`**: An array containing the distances between the query vector and the 3 nearest vectors in the index.
  - For example: `D = [[0.25, 0.35, 0.5]]`
    - The first number (0.25) is the distance between the query vector and the first nearest neighbor.
    - The second number (0.35) is the distance to the second nearest neighbor.
    - The third number (0.5) is the distance to the third nearest neighbor.

- **`I`**: An array containing the indices of the 3 nearest neighbors in the index.
  - For example: `I = [[5, 8, 3]]`
    - The first value (5) refers to the index of the closest vector in the FAISS index.
    - The second value (8) refers to the second closest, and so on.

You can then use `I` to look up the actual data associated with the nearest vectors from your dataset (e.g., the text or images associated with those vectors). The distances in `D` tell you how similar those vectors are to your query vector.
